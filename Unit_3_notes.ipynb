````
# A class is a blueprint for an instance of that class.
class Employee:
  pass

emp_1 = Employee()
emp_2 = Employee()

# Locations are different, they are not the same thing.
print(emp_1)
print(emp_2)
````

````
# A class is a blueprint for an instance of that class.
class Employee:
  
  def __init__(self, first, last, pay):
    # instance variables
    self.first = first
    self.last = last
    self.pay = pay
    self.email = first + '.' + last + '@email.com'

emp_1 = Employee('John', 'Doe', 85000)
emp_2 = Employee('Beth', 'Sew', 90000)

# Reach into the employees and get the emails
print(emp_1.email)
print(emp_2.email)

# Get emp_1 full name THE HARD WAY
print('{} {}'.format(emp_1.first, emp_1.last)) # OR
print(emp_1.first + ' ' + emp_1.last)
````

````
# A class is a blueprint for an instance of that class.
class Employee:

  # __init__ method --> Other languages call it the constructor
  def __init__(self, first, last, pay):
    # instance variables
    self.first = first
    self.last = last
    self.pay = pay
    self.email = first + '.' + last + '@email.com'

  # Method to make getting full name EASIER
  def full_name(self):
    return self.first + ' ' + self.last

# INSTANTIATE employees
emp_1 = Employee('John', 'Doe', 85000)
emp_2 = Employee('Beth', 'Sew', 90000)

# Get emp_1 full name THE HARD WAY
print("# Get emp_1 full name THE HARD WAY\n---") 

print("print('{} {}'.format(emp_1.first, emp_1.last))")
print('{} {}'.format(emp_1.first, emp_1.last)) # OR

print("print(emp_1.first + ' ' + emp_1.last)")
print(emp_1.first + ' ' + emp_1.last)

# Seperate things
print('---\n')

# Get emp_1 full name THE EASY WAY
print('Get emp_1 full name THE EASY WAY')
print("print(emp_2.full_name())")
print(emp_2.full_name())
````

**As a program grows, it's easy to see how writing the code in the `print(emp_2.full_name())` way is faster.**

````
# Introducing a class variable

class Employee:

  # CLASS VARIABLES
  raise_amount = 1.04

  def __init__(self, first, last, pay):
    # INSTANCE VARIABLES
    self.first = first
    self.last = last
    self.pay = pay
    self.email = first + '.' + last + '@email.com'

  # Method to make full name
  def full_name(self):
    return self.first + ' ' + self.last
  
  # Introducing a need for CLASS VARIABLE
  def apply_raise(self):
    self.pay = int(self.pay * self.raise_amount)

# INSTANTIATE employees
emp_1 = Employee('John', 'Doe', 85000)
emp_2 = Employee('Beth', 'Sew', 90000)

# Get emp_1&2 pay
print(emp_1.pay)
print(emp_2.pay)

# APPLY A RAISE TO BOTH
emp_1.apply_raise()
emp_2.apply_raise()

# New line
print('\n')

# Check that the raise was applied
print(emp_1.pay)
print(emp_2.pay)
````

**Unit Test Basic Example** from [here](https://docs.python.org/3.4/library/unittest.html)

* Make test folder (outside everything else, but inside the repo-root)
* Create `__init__.py` file
* Create test file named after the `.py` file to test. **MUST END IN `_test.py`**
* Import `import unittest` along with the methods to be tested, (ex. `from my_lambdata.my_mod import ten_x`)
* Make Class `TestWhatever`
* Make Class methods **MUST START WITH `test_`** and pass `self`
* Use `self.assert*` for testing ([check numpy test support](https://docs.scipy.org/doc/numpy/reference/routines.testing.html))
* Add  
```
if __name__ == '__main__':
    unittest.main()
    ```
* Run test with command like `python -m test.my_mod_test` where `test` is the folder (because we've pipenv'd to the repo-root) and `my_mod_test` is the file name

```
import unittest

class TestStringMethods(unittest.TestCase):

  def test_upper(self):
      self.assertEqual('foo'.upper(), 'FOO')

  def test_isupper(self):
      self.assertTrue('FOO'.isupper())
      self.assertFalse('Foo'.isupper())

  def test_split(self):
      s = 'hello world'
      self.assertEqual(s.split(), ['hello', 'world'])
      # check that s.split fails when the separator is not a string
      with self.assertRaises(TypeError):
          s.split(2)

if __name__ == '__main__':
    unittest.main()
```
